```python
import numpy as np
import common as c
dtype = np.int
```

```python
def next_sylvester(M):
    return np.hstack([np.vstack([M, M]), np.vstack([M, -M])])
```

```python
def sylvester_bin(M):
    B = np.copy(M)
    B[B == 1] = 0
    B[B == -1] = 1
    return B
```

```python
H1 = np.array([[1]], dtype=dtype)
c.display(H1)
```

```python
H2 = next_sylvester(H1)
c.display(H2)
```

```python
H3 = next_sylvester(H2)
c.display(H3)
```

```python
H4 = next_sylvester(H3)
c.display(H4)
```

```python
B4 = sylvester_bin(H4)
c.display(B4)
```

```python
B4rref = c.deletez(c.brref(B4))
c.display(B4rref)
```
