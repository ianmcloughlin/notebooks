# Big-O Notation

***

Let $f$ and $g$ be functions $f: \mathbb{N} \rightarrow \mathbb{R}^+$ and $g: \mathbb{N} \rightarrow \mathbb{R}^+$.
Say that $f(n) = O(g(n))$ if positive integers $c$ and $n_0$ exist such that for every integer $n \ge n_0$ then

$$f(n) \le c g(n)$$

$\mathbb{N}$ is the set of natural numbers.

$\mathbb{R}^+$ is the set of positive real numbers.

## Imports

***

```python
# Numerical arrays.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt
```

## Example

***

$a(n) = n^2$

$b(n) = n \log n$


### $n^2$

```python
# Range of values for n.
n = np.arange(1, 2001)

# Show.
n
```

```python
# a(n)
a = n**2

# Show.
a
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, a, 'b.', label='$n^2$')

# Legend.
ax.legend();
```

### $\log_{10} n$

```python
# log(n)
log = np.log10(n)

# Show.
log
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot log(n).
ax.plot(n, log, 'g.', label='$\log_{10} n$')

# Legend.
ax.legend();
```

### $n \log_{10} n$

***

```python
# Range of values for n.
n = np.arange(1, 2001)

# Show.
n
```

```python
# b(n)
b = n * np.log10(n)

# Show.
b
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot b(n).
ax.plot(n, b, 'r.', label='$n \log_{10} n$')

# Legend.
ax.legend();
```

### $a(n)$ and $b(n)$

***

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, a, 'b.', label='$n^2$')

# Plot b(n).
ax.plot(n, b, 'r.', label='$n \log_{10} n$')

# Legend.
ax.legend();
```
