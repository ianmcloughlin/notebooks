# Big-O Notation

***

> Let $f$ and $g$ be functions $f: \mathbb{N} \rightarrow \mathbb{R}^+$ and $g: \mathbb{N} \rightarrow \mathbb{R}^+$.
> 
> Say that $f(n) = O(g(n))$ if positive integers $c$ and $n_0$ exist such that for every integer $n \ge n_0$ then
>
> $f(n) \le c g(n)$
> 
> &mdash; Introduction to the Theory of Computation, 3rd Edition, Michael Sipser, p.248.

$\mathbb{N}$ is the set of natural numbers $\{1, 2, 3, \ldots \}$.

$\mathbb{R}^+$ is the set of positive real numbers, the positive side of the real number line.

## Imports

***

```python
# Numerical arrays.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt
```

## Example

***

$a(n) = n^2$

$b(n) = n \log n$


```python
# Range of values for n.
n = np.arange(1, 21)

# Show.
n
```

### $a(n) = n^2$

```python
# a(n)
a = lambda n: n**2

# Show.
a(n)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, a(n), label='$n^2$', color='blue', marker='.', linestyle='None')

# Legend.
ax.legend();
```

### $\log_{10} n$

```python
# log(n)
log = lambda n: np.log10(n)

# Show.
log(n)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot log(n).
ax.plot(n, log(n), label='$\log_{10} n$', color='green', marker='.', linestyle='None')

# Legend.
ax.legend();
```

### $b(n) = n \log_{10} n$

***

```python
# b(n)
b = lambda n: n * np.log10(n)

# Show.
b(n)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot b(n).
ax.plot(n, b(n), label='$n \log_{10} n$', color='red', marker='.', linestyle='None')

# Legend.
ax.legend();
```

### $a(n)$ and $b(n)$

***

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, a(n), label='$n^2$', color='blue', marker='.', linestyle='None')

# Plot b(n).
ax.plot(n, b(n), label='$n \log_{10} n$', color='red', marker='.', linestyle='None')

# Legend.
ax.legend();
```
