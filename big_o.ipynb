# Big-O Notation

***

> Let $f$ and $g$ be functions $f: \mathbb{N} \rightarrow \mathbb{R}^+$ and $g: \mathbb{N} \rightarrow \mathbb{R}^+$.
> 
> Say that $f(n) = O(g(n))$ if positive integers $c$ and $n_0$ exist such that for every integer $n \ge n_0$ then
>
> $f(n) \le c g(n)$
> 
> &mdash; Introduction to the Theory of Computation, 3rd Edition, Michael Sipser, p.248.

$\mathbb{N}$ is the set of natural numbers $\{1, 2, 3, \ldots \}$.

$\mathbb{R}^+$ is the set of positive real numbers, the positive side of the real number line.

## Imports

***

```python
# Numerical arrays.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt

# Math.
import math
```

## Example

***

$a(n) = n^2$ and $b(n) = n \log n$


```python
# Range of values for n.
n = np.arange(1, 21)

# Show.
n
```

$a(n) = n^2$

```python
# a(n)
a = lambda n: n**2

# Show.
a(n)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, a(n), label='$n^2$', color='blue', marker='.', linestyle='None')

# Legend.
ax.legend();
```

$\log_{10} n$

$x^y = z \Leftrightarrow \log_x z = y$

```python
# log(n)
log = lambda n: np.log10(n)

# Show.
log(n)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot log(n).
ax.plot(n, log(n), label='$\log_{10} n$', color='green', marker='.', linestyle='None')

# Legend.
ax.legend();
```

$b(n) = n \log_{10} n$

```python
# b(n)
b = lambda n: n * np.log10(n)

# Show.
b(n)
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot b(n).
ax.plot(n, b(n), label='$n \log_{10} n$', color='red', marker='.', linestyle='None')

# Legend.
ax.legend();
```

$n^2$ and $n\log{n}$

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, a(n), label='$n^2$', color='blue', marker='.', linestyle='None')

# Plot b(n).
ax.plot(n, b(n), label='$n \log_{10} n$', color='red', marker='.', linestyle='None')

# Legend.
ax.legend();
```

$n \log{n}$, $n^2$ and $2^n$

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot nlogn.
ax.plot(n, n * np.log10(n), label='$n \log_{10} n$', color='red', marker='.', linestyle='None')

# Plot n^2.
ax.plot(n, n**2, label='$n^2$', color='blue', marker='.', linestyle='None')

# Plot 2^n.
ax.plot(n, 2**n, label='$2^n$', color='green', marker='.', linestyle='None')

# Legend.
ax.legend();
```

$n \log{n}$, $n^2$, $2^n$, and $n!$

```python
# Factorial function for numpy arrays.
fact = np.vectorize(math.factorial)
```

```python
# Range of values for n - 20 too big for factorial.
n = np.arange(1, 11)

# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot nlogn.
ax.plot(n, n * np.log10(n), label='$n \log_{10} n$', color='red', marker='.', linestyle='None')

# Plot n^2.
ax.plot(n, n**2, label='$n^2$', color='blue', marker='.', linestyle='None')

# Plot 2^n.
ax.plot(n, 2**n, label='$2^n$', color='green', marker='.', linestyle='None')

# Plot n!
ax.plot(n, fact(n), label='$n!$', color='purple', marker='.', linestyle='None')

# Legend.
ax.legend();
```

## The Constants

***

$n \ge n_0$

$n^3$ is $O(2^n)$ with $n_0 = 10$

```python
# Range of values for n.
n = np.arange(1, 13)

# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, n**3, label='$n^3$', color='blue', marker='.', linestyle='None')

# Plot b(n).
ax.plot(n, 2**n, label='$2^n$', color='red', marker='.', linestyle='None')

# Legend.
ax.legend();
```

$n^2 + n + 20$ is $O(n^2)$ with $c = 3$

The biggest term dominates so replace all terms with it: $n^2 + n + 20 \rightarrow n^2 + n^2 + n^2 = 3n^2$

![Why does the largest term not simply eat the other terms?](img/eattheotherterms.png)

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(10, 4))

# Plot a(n).
ax.plot(n, n**2, label='$n^2$', color='blue', marker='.', linestyle='None')

# Plot b(n).
ax.plot(n, n**2 + n + 20, label='$n^2 + n + 20$', color='black', marker='.', linestyle='None')

# Legend.
ax.legend();
```
