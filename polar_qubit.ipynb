```python
import matplotlib.pyplot as plt

import numpy as np

import qiskit

plt.style.use('./ian.mplstyle')
```

```python
def qubit(c0, c1):
  return qiskit.quantum_info.Statevector([c0, c1])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.add_patch(plt.Circle((0.0, 0.0), 1, linestyle=':', fill=None))

ax.axhline(0.0, color='grey', linestyle='-')
ax.axvline(0.0, color='grey', linestyle='-')

ax.set_xlim(-2.0, 2.0)
ax.set_ylim(-2.0, 2.0)

q0 = qubit(1.0/np.sqrt(2) + 0.0j, -1.0/np.sqrt(2) + 0.0j)

ax.plot([0.0, q0[0].real], [0.0, q0[0].imag], 'm-')
ax.plot([0.0, q0[0].real], [0.0, q0[0].imag], 'k.')

ax.plot([0.0, q0[1].real], [0.0, q0[1].imag], 'm-')
ax.plot([0.0, q0[1].real], [0.0, q0[1].imag], 'k.')

ax.set_aspect("equal");
```
