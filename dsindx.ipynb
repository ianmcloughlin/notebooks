```python
import matplotlib.pyplot as plt

import numpy as np

plt.style.use('./ian.mplstyle')
```

```python
x = np.linspace(-3.0 * np.pi, 4.0 * np.pi, 1000)

y = np.sin(x)
```

```python
point_x = 1.0
point_y = np.sin(point_x)
m = np.cos(point_x)
c = point_y - m * point_x

dx_left = 5.0
dx_right = 1.0

line_x = np.linspace(point_x - dx_left, point_x + dx_right, 10)
line_y = m * line_x + c
```

```python
fig, ax = plt.subplots(figsize=(10, 2))

ax.plot(x, y, 'purple')

ax.plot(line_x, line_y)

ax.set_aspect("equal");
```
