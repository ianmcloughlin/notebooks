## Div, Grad, Curl, and All that (Schey)

```python
N = 10
x, y = np.meshgrid(np.arange(-N, N), np.arange(-N, N))
```

```python
# Problem 1.1 (a): iy + jx
i, j = y, x

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=100, angles='xy');
```

```python
# Problem 1.1 (b): (i + j)/sqrt(2)
i, j = 0 * x + 1.0 / np.sqrt(2), 0 * y + 1.0 / np.sqrt(2)

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=40, angles='xy');
```

```python
# Problem 1.1 (c): ix - jy
i, j = x, -y

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=100, angles='xy');
```

```python
# Problem 1.1 (d): iy
i, j = y, 0 * x

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=200, angles='xy');
```

```python
# Problem 1.1 (e): jx
i, j = 0 * x, x

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=200, angles='xy');
```

```python
# Problem 1.1 (f): (iy + jx)/sqrt(x^2 + y^2)
mag = np.sqrt(x**2 + y**2)
i, j = y / mag, x / mag

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=50, angles='xy');
```

```python
# Problem 1.1 (g): iy + jxy
i, j = y, x * y

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=500, angles='xy');
```

```python
# Problem 1.1 (h): i + jy
i, j = 0.0 * x + 1.0, y

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=50, angles='xy');
```

```python
# Problem 1.2
e_0 = 8.854 #e-12
E = np.divide(1.0 / (4.0 * np.pi *e_0), np.sqrt(x**2 + y**2), out=np.zeros_like(x, dtype=float), where=np.sqrt(x**2 + y**2)!=0)

i, j = -np.sign(x) * E, -np.sign(y) * E

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=0.2, angles='xy');
```

## Problem 1.4 (c)

$ r = ix + jy = i a \cos (\omega t) + j b \sin (\omega t) $

$\forall x: \cos^2 x + \sin^2 x = 1$

$\Rightarrow \cos^2 (\omega t) + \sin^2 (\omega t) = 1$

$\forall x: x^2 / x^2 = 1$

$\Rightarrow (a^2 / a^2) \cos^2 (\omega t) + (b^2 / b^2) \sin^2 (\omega t) = 1$

$\Rightarrow (a^2 \cos^2 (\omega t)) / a^2 + (b^2 \sin^2 (\omega t)) / b^2 = 1$

$\Rightarrow (a \cos (\omega t))^2 / a^2 + (b \sin (\omega t))^2 / b^2 = 1$

$\Rightarrow x^2 / a^2 + y^2 / b^2 = 1$

$\Rightarrow (x/a)^2 + (y/b)^2 = 1$

## Problem 1.5

$$E(r) = \frac{1}{4 \pi \epsilon_0} \sum_{t=1}^N \frac{q_t}{\vert r - r_t \vert^2} \hat{u}_t$$

***

$q_1 = 1$

$q_2 = -1$

$\vert r - r_1 \vert^2 = \vert (0, y, 0) - (1, 0, 0) \vert^2 = (-1)^2 + (y)^2 + (0^2) = 1 + y^2$

$\vert r - r_2 \vert^2 = \vert (0, y, 0) - (-1, 0, 0) \vert^2 = (1)^2 + (y)^2 + (0^2) = 1 + y^2$

$\hat{u}_1 = (0i + yj + 0k - 1i - 0j - 0k) / \sqrt{1 + y^2} = (-1i + yj) / \sqrt{1 + y^2}$

$\hat{u}_2 = (0i + yj + 0k - (-1)i - 0j - 0k) / \sqrt{1 + y^2} = (1i + yj) / \sqrt{1 + y^2}$

$\Rightarrow E(r) = (4 \pi \epsilon_0)^{-1} \left[ (1 / (1 + y^2)) ((-1i + yj) / \sqrt{1 + y^2}) + (-1 / (1 + y^2)) ((1i + yj) / \sqrt{1 + y^2}) \right]$

$\Rightarrow E(r) = (4 \pi \epsilon_0)^{-1} \left[ (1 / (1 + y^2)) ((-1i + yj) / \sqrt{1 + y^2}) - (1 / (1 + y^2)) ((1i + yj) / \sqrt{1 + y^2}) \right]$

$\Rightarrow E(r) = (4 \pi \epsilon_0)^{-1} (1 + y^2)^{-1} (\sqrt{1 + y^2})^{-1} \left[ (-1i + yj)  - (1i + yj) \right]$

$\Rightarrow E(r) = (4 \pi \epsilon_0)^{-1} (1 + y^2)^{-1} (\sqrt{1 + y^2})^{-1} (-2i)$

$\Rightarrow E(r) = (2 \pi \epsilon_0)^{-1} (1 + y^2)^{-3/2} (-i)$
